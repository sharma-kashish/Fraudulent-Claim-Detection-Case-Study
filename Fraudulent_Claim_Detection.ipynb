#### **4.4.1** Target likelihood analysis for categorical variables. <font color="red">[5 Marks]</font>
Investigate the relationships between categorical features and the target variable by analysing the target event likelihood (for the `'Y'` event) for each level of every relevant categorical feature. Through this analysis, identify categorical features that do not contribute much in explaining the variation in the target variable.

```python
# Write a function to calculate and analyse the target variable likelihood for categorical features

def target_likelihood_analysis(df, target_col):
    # Filter categorical columns
    cat_cols = df.select_dtypes(include='object').columns.tolist()
    if target_col in cat_cols:
        cat_cols.remove(target_col)

    # Convert target to binary if it's 'Y'/'N'
    if df[target_col].dtype == 'object':
        df['target_col_orignal'] = df[target_col]
        df[target_col] = df[target_col].map({'Y': 1, 'N': 0})

    # Set plotting style
    sns.set(style="whitegrid")

    for col in cat_cols:
        # Calculate fraud likelihood by category
        likelihood_df = pd.pivot_table(data=df,index=col,columns='target_col_orignal', values=target_col ,aggfunc='count')
        # Sort columns so 'Y' comes before 'N'
        likelihood_df = likelihood_df.sort_index(axis=1, ascending=False)

        #Let's normalize the pivot for each row
        normalized_values = likelihood_df.div(likelihood_df.sum(axis=1), axis=0)
        # print(normalized_values)
        print("\n Normalized Values for {} with respect to target column - {}\n".format(col,target_col))
        print(normalized_values)

        # Plot
        plt.figure(figsize=(4, 6))
        normalized_values.plot(kind='bar',stacked=True)
        plt.title(f"Fraud Likelihood by {col}")
        plt.ylim(0, 1)
        plt.ylabel("Likelihood of Fraud")
        plt.xlabel(col)
        plt.xticks(rotation=90)
        plt.tight_layout()
        plt.show()
```

```python
#Let's concatinate the X_train and y_train into single data frame
train = pd.concat([X_train, y_train], axis=1)

# Call function to see the results
target_likelihood_analysis(train, y_train.columns[0])  # here y_train.columns[0] is the name of target column

#let's drop target_col_orignal which we created for pivot generation
df = df.drop(columns='target_col_orignal')
```

### **6.4 Combine values in Categorical Columns** <font color="red">[6 Marks]</font>
During the EDA process, categorical columns with multiple unique values may be identified. To enhance model performance, it is essential to refine these categorical features by grouping values that have low frequency or provide limited predictive information.

Combine categories that occur infrequently or exhibit similar behavior to reduce sparsity and improve model generalisation.

```python
# Combine categories that have low frequency or provide limited predictive information

# Identify categorical columns (excluding target)
cat_cols = X_train.select_dtypes(include=['object']).columns

# Dictionary to store value mappings for each categorical column
category_mappings = {}

# Function to check if a value should be treated as missing/unknown
def is_unknown(val):
    return val in ['?', 'None', None, np.nan]

# Loop through each categorical column
for col in cat_cols:
    # Get value counts in training data
    value_counts = X_train[col].value_counts()
    total_count = len(X_train)
    
    # Define threshold for rare categories (less than 2% of data)
    rare_threshold = total_count * 0.02
    
    # Identify rare categories
    rare_categories = value_counts[value_counts < rare_threshold].index.tolist()
   
    if rare_categories:
        # Create mapping for this column
        mapping = {category: 'Other' if category in rare_categories else category 
                  for category in X_train[col].unique()}
        
        # Handle missing/unknown values consistently
        for unknown_val in ['?', 'None', None, np.nan]:
            mapping[unknown_val] = 'Unknown'
            
        category_mappings[col] = mapping
        
        # Apply mapping to train data
        X_train[col] = X_train[col].map(lambda x: 'Unknown' if is_unknown(x) else mapping.get(x, 'Other'))
        
        # Apply mapping to test data, handling any new categories in test
        X_test[col] = X_test[col].map(lambda x: 'Unknown' if is_unknown(x) else mapping.get(x, 'Other'))
        
        # Count before and after to show reduction in categories
        before_count = len(value_counts)
        after_count = X_train[col].nunique()
        # Print details about the column if rare category found
        print(f"\nColumn '{col}'")
        print(f"  Total unique values: {X_train[col].nunique()}")
        print(f"  Rare categories (< 2% of data): {len(rare_categories)}")
        print(f"  Reduced categories from {before_count} to {after_count}")
        print(f"  Top 5 categories after combining: {X_train[col].value_counts().nlargest(5).to_dict()}")
    else:
        # Still handle missing/unknown values consistently even if no rare categories
        X_train[col] = X_train[col].map(lambda x: 'Unknown' if is_unknown(x) else x)
        X_test[col] = X_test[col].map(lambda x: 'Unknown' if is_unknown(x) else x)

columns_modified = len(category_mappings)
print(f"\nModified {columns_modified} categorical columns by combining rare categories in both Train and Test dataset")

```

### **7.3 Find the Optimal Cutoff** <font color = red>[12 marks]</font>

#### **7.3.1** Plot ROC Curve  to visualise the trade-off between true positive rate and false positive rate across different classification thresholds <font color="red">[2 Marks]</font>

```python
# Import libraries or function to plot the ROC curve
from sklearn.metrics import roc_curve, auc, roc_auc_score
import matplotlib.pyplot as plt


# Define ROC function
def plot_roc_curve(y_true, y_pred_prob):
    fpr, tpr, thresholds = roc_curve(y_true, y_pred_prob)
    auc_score = roc_auc_score(y_true, y_pred_prob)

    plt.figure(figsize=(8, 5))
    plt.plot(fpr, tpr, color='darkorange', lw=2, label=f'ROC curve (AUC = {auc_score:.4f})')
    plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate (Recall)')
    plt.title('Receiver Operating Characteristic (ROC) Curve')
    plt.legend(loc='lower right')
    plt.grid(True)
    plt.show()
	
```

**Sensitivity and Specificity tradeoff**

After analysing the area under the curve of the ROC, check the sensitivity and specificity tradeoff to find the optimal cutoff point.

```python
# Define function to calculate metrics for multiple cutoff values
def evaluate_cutoffs(y_true, y_prob):
    cutoffs = np.arange(0.1, 0.9, 0.05)
    sensitivity = []
    specificity = []
    accuracy = []

    for cutoff in cutoffs:
        predicted_class = (y_prob >= cutoff).astype(int)
        tn, fp, fn, tp = metrics.confusion_matrix(y_true, predicted_class).ravel()
        
        sensitivity.append(tp / (tp + fn))
        specificity.append(tn / (tn + fp))
        accuracy.append((tp + tn) / (tp + tn + fp + fn))

    return pd.DataFrame({
        'cutoff': cutoffs,
        'sensitivity': sensitivity,
        'specificity': specificity,
        'accuracy': accuracy
    })

# Get metrics across cutoffs
cutoff_df = evaluate_cutoffs(y_train, predictions_df['predicted_proba'])

# Plot the metrics
plt.figure(figsize=(20, 6))
plt.plot(cutoff_df['cutoff'], cutoff_df['sensitivity'], label='Sensitivity', color='green')
plt.plot(cutoff_df['cutoff'], cutoff_df['specificity'], label='Specificity', color='blue')
plt.plot(cutoff_df['cutoff'], cutoff_df['accuracy'], label='Accuracy', color='red')
plt.xlabel('Cutoff Threshold')
plt.ylabel('Score')
plt.title('Sensitivity, Specificity, and Accuracy at Different Cutoffs')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Compute the absolute difference between sensitivity and specificity
cutoff_df['diff'] = abs(cutoff_df['sensitivity'] - cutoff_df['specificity'])

# Find the row where the difference is smallest
optimal_cutoff_row = cutoff_df.loc[cutoff_df['diff'].idxmin()]
optimal_cutoff = optimal_cutoff_row['cutoff']

# Display the optimal cutoff and associated metrics
print(f"Optimal Cutoff: {optimal_cutoff:.2f}")
print(f"Sensitivity: {optimal_cutoff_row['sensitivity']:.4f}")
print(f"Specificity: {optimal_cutoff_row['specificity']:.4f}")
print(f"Accuracy: {optimal_cutoff_row['accuracy']:.4f}")
```
